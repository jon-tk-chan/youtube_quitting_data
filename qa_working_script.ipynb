```python
filename = "/content/sample_data/processed_ALL.json"
with open(filename, 'r') as f:
  data = json.load(f)

data.keys()
all_data = list(data.items())[:500]
all_data[0]
```

```python

i = 0
questions= [
    'What was his or her job?',
    'Why did they quit?',
    'What is the main reason for him or her to quit?'
]

to_df = {
    "id": [],
    "prev_job": [],
    "quit_rsn1": [],
    "quit_rsn2": []
    
}

for i, item in enumerate(all_data):
  id = item[0]
  context = item[1]['str']

  qa_0 = nlp({
    'question': questions[0],
    'context': context
  })

  qa_1 = nlp({
    'question': questions[1],
    'context': context
  })

  qa_2 = nlp({
    'question': questions[2],
    'context': context
  })

  print(id)
  # print(qa_0['answer'])
  # print(qa_1['answer'])
  # print(qa_2['answer'])

  prev_job = qa_0['answer']
  reason1 = unidecode.unidecode(qa_1['answer'])
  reason2= unidecode.unidecode(qa_2['answer'])
  to_df['id'].append(id)
  to_df['prev_job'].append(prev_job)
  to_df['quit_rsn1'].append(reason1)
  to_df['quit_rsn2'].append(reason2)
  # i +=1

  if i % 50 == 0:
    print("PROCESSED: ", i)
  # if i == 5:
  #   break 

out_df = pd.DataFrame(to_df)
out_df.head()

out_df.to_csv("/content/drive/MyDrive/youtube_output/quitting_qa_out.csv")
```
